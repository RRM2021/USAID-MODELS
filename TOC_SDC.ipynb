```python
query = """
SELECT t1.NOMBRE_PROB, t1.NOMBRE_CAUSAS, t3.NOMBRE_OBJETIVO
FROM N t1
JOIN M t2 ON t1.NOMBRE_CAUSAS = t2.NOMBRE_CAUSAS
JOIN O t3 ON t1.NOMBRE_CAUSAS = t3.NOMBRE_CAUSAS
WHERE NOMBRE_PROB = "El conflicto armado"
ORDER BY t1.NOMBRE_PROB, t1.NOMBRE_CAUSAS, t3.NOMBRE_OBJETIVO
"""

result = spark.sql(query)
result.show(result.count(), truncate=False)
```

```python
query = """
SELECT t1.NOMBRE_PROB, t2.NOMBRE_POT, t1.NOMBRE_CAUSAS, t3.NOMBRE_OBJETIVO, t4.DO_ID
FROM N t1
JOIN M t2 ON t1.NOMBRE_CAUSAS = t2.NOMBRE_CAUSAS
JOIN O t3 ON t1.NOMBRE_CAUSAS = t3.NOMBRE_CAUSAS
JOIN P t4 ON t4.NOMBRE_OBJETIVO = t3.NOMBRE_OBJETIVO
WHERE NOMBRE_PROB = "El conflicto armado"
ORDER BY t1.NOMBRE_PROB, t2.NOMBRE_POT, t1.NOMBRE_CAUSAS, t3.NOMBRE_OBJETIVO
"""

result = spark.sql(query)
result.show()
```

```python
from pyspark.sql.functions import col

# Count rows for each column
column_counts = [result.select(col(column)).count() for column in result.columns]

# Create a dictionary with column names and their respective row counts
column_counts_dict = dict(zip(result.columns, column_counts))

# Print the row counts for each column
for column, count in column_counts_dict.items():
    print(f"Column '{column}': {count} rows")
```
